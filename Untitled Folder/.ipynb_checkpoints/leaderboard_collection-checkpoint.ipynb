```python
feature_cols=['wins','losses','hotstreak']

testx = X[feature_cols]
y= X.leaguePoints
```

```python
# instantiate a DecisionTreeRegressor (with random_state=1)
from sklearn.tree import DecisionTreeRegressor
treereg = DecisionTreeRegressor(random_state=1)
treereg
```

```python
# use leave-one-out cross-validation (LOOCV) to estimate the RMSE for this model
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(treereg, testx, y, cv=14, scoring='mean_squared_error')
np.mean(np.sqrt(-scores))
```

```python
# try different values one-by-one
treereg = DecisionTreeRegressor(max_depth=1, random_state=1)
scores = cross_val_score(treereg, testx, y, cv=14, scoring='mean_squared_error')
np.mean(np.sqrt(-scores))
```

```python
# list of values to try
max_depth_range = range(1, 8)
```

```python
# list to store the average RMSE for each value of max_depth
RMSE_scores = []
```

```python
# use LOOCV with each value of max_depth
for depth in max_depth_range:
    treereg = DecisionTreeRegressor(max_depth=depth, random_state=1)
    MSE_scores = cross_val_score(treereg, testx, y, cv=14, scoring='mean_squared_error')
    RMSE_scores.append(np.mean(np.sqrt(-MSE_scores)))
```

```python
# plot max_depth (x-axis) versus RMSE (y-axis)
plt.plot(max_depth_range, RMSE_scores)
plt.xlabel('max_depth')
plt.ylabel('RMSE (lower is better)')
```

```python
# max_depth=3 was best, so fit a tree using that parameter
treereg = DecisionTreeRegressor(max_depth=3, random_state=1)
treereg.fit(testx, y)
```

```python
# "Gini importance" of each feature: the (normalized) total reduction of error 
# brought by that feature
pd.DataFrame({'feature':feature_cols, 'importance':treereg.feature_importances_})
```
