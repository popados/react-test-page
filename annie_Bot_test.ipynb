```python
import json
import urllib2
from urllib2 import Request, urlopen
import pandas as pd
import csv
from pandas.io.json import json_normalize
import numpy as np
import matplotlib.pyplot as plt
import unicodedata




request = Request('https://na.api.pvp.net/api/lol/na/v2.2/matchlist/by-summoner/35590582?beginIndex=0&endIndex=10&api_key=RGAPI-6b4e9139-2121-4820-8cd9-371891441c4b')

response = urlopen(request)

data = response.read()

matchList= json.loads(data)

matchList
```

```python
df_cols =['champion','lane','matchID','platformID','queue','region','role','season','timestamp']

champion=[]
lane=[]
matchId=[]
platformId=[]
queue=[]
region=[]
role=[]
season=[]
timestamp=[]

for matches in matchList['matches']:    
    champion.append(matches[u'champion'])
    lane.append(matches[u'lane'])
    matchId.append(matches[u'matchId'])
    platformId.append(matches[u'platformId'])
    queue.append(matches[u'queue'])
    region.append(matches[u'region'])
    role.append(matches[u'role'])
    season.append(matches[u'season'])
    timestamp.append(matches[u'timestamp'])
    
df= pd.DataFrame([champion,lane,matchId,platformId,queue,region,role,season,timestamp]).T

```

```python
# df

with open('matchList1.csv', 'wb') as createcsv:
    create = csv.writer(createcsv)
    create.writerow(df_cols)
    df.to_csv('matchList1.csv', index =False, encoding="utf-8")
    
```

```python
print loadMatch1['teams']

#{'matchid'},
team_col_header = ['match','firstDragon','bans','firstInhibitor','baronKills','firstRiftHerald','winner','firstBaron','riftheraldKills',
                  'firstBlood','teamId','firstTower','vilemawKills','inhibitorKills','towerKills','dominionVictoryScore',
                  'dragonKills']

#matchid=['2405647360']
match=[]
firstDragon=[]
bans=[]
firstInhibitor=[]
baronKills=[]
firstRiftHerald=[]
winner=[]
firstBaron=[]
riftHeraldKills=[]
firstBlood=[]
teamId=[]
firstTower=[]
vilemawKills=[]
inhibitorKills=[]
towerKills=[]
dominionVictoryScore=[]
dragonKills=[]

for teams in loadMatch1['teams']:
    firstDragon.append(teams[u'firstDragon'])
    bans.append(teams[u'bans'])
    firstInhibitor.append(teams[u'firstInhibitor'])
    baronKills.append(teams[u'baronKills'])
    firstRiftHerald.append(teams[u'firstRiftHerald'])
    winner.append(teams[u'winner'])
    firstBaron.append(teams[u'firstBaron'])
    riftHeraldKills.append(teams[u'riftHeraldKills'])
    firstBlood.append(teams[u'firstBlood'])
    teamId.append(teams[u'teamId'])
    firstTower.append(teams[u'firstTower'])
    vilemawKills.append(teams[u'vilemawKills'])
    inhibitorKills.append(teams[u'inhibitorKills'])
    towerKills.append(teams[u'towerKills'])
    dominionVictoryScore.append(teams[u'dominionVictoryScore'])
    dragonKills.append(teams[u'dragonKills'])
    
dfMatch = pd.DataFrame([firstDragon,bans,firstInhibitor,baronKills,firstRiftHerald,winner,firstBaron,riftHeraldKills,
                  firstBlood,teamId,firstTower,vilemawKills,inhibitorKills,towerKills,dominionVictoryScore,
                  dragonKills]).T
    
```
